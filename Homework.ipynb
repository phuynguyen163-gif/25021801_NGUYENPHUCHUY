```python
#Bài 6
import math
a, b, c=map(float,input().split())
p=(a+b+c)/2
s=math.sqrt(p*(p-a)*(p-b)*(p-c))
if a**2==b**2+c**2:
  print("diện tích tam giác là:",s)
elif b**2==a**2+c**2:
  print("diện tích tam giác là:",s)
elif c**2==a**2+b**2:
  print("diện tích tam giác là:",s)
else:
  print("không phải 3 cạnh của tam giác")
```
